```python
import pandas as pd

df = pd.read_csv(r"C:\Users\DELL\Downloads\Titanic-Dataset.csv")

df.head()
```

```python
filtered_df = df[(df['Age'] >= 18) & (df['Fare'] > 30)]
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1
df['Fare_Per_Person'] = df['Fare'] / df['Family_Size']
summary = df.groupby(['Pclass', 'Sex']).agg({
    'Fare': 'mean',
    'Survived': 'mean'
}).rename(columns={
    'Fare': 'Average_Fare',
    'Survived': 'Survival_Rate'
})
print("Filtered Passengers:")
print(filtered_df.head())
print("\nSummary Analysis:")
print(summary)
df.to_csv("task5_processed_titanic.csv", index=False)
```
